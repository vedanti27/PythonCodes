```python
# 2.a. Determine how many missing values there are in each column of the dataset
#Can you think of a reason why these values are missing? Discuss different strategies for filling the missing
#values, highlighting the advantages and disadvantages of each strategy, in the context of this
#dataset. 

# Explanation:
#.  Reasons for missing values or data loss:
#     1. Missing data could be due to incomplete data entry by the system or user.
#     2. Equipment malfunctions due to which values could not be recorded.
#     3. Lost values or files before data was stored.


#.  Strategies for filling the missing values 
#         1. fillna()- We can mention a value that we want to fill for missing data.
#                      We can specify any statistics function as well.
#                     Advantage - Useful in real world as we can give realistic values as per the dataset 
#                                requirement.
#                                 Simple to use, pandas function. Replacement can be controlled.
#                     Disadvantages - need to provide values for stats function.
                    
#         2. SimpleImputer - It adds missing values calculated by any stat function(mean, median or mode) 
#                            Helpful in predictive analysis.
#                 Advantage - Missing value featurewise, useful when there is really big dataset, Better speed.
#                 Disadvantages - Does not consider dependencies among other attributes values in dataset.
#                                 No control over replacement values
                
#         3. IterativeImputer - missing values of one feature is internally calculated by a model that fits 
                                #on other 
#                               features,this happens iteratively for all the features. 
#             Advantages - calculates model based on multiple features, gives better result than SimpleImputer 
                          #and fillna
#             Disadvantages - Time-consuming as it calculates values for all features.
            
#         4. KNNImputer  - KNN algorithm is used to fill missing values of a particular feature.
#            Advantages - Generally, it is less time consuming than IterativeImputer.
#            Disadvantages- Not good for large dataset as it stores whole dataset in memory, 
#                           specify the number of neighbours, senstive to outliers.
           

#Code:
print ('Missing value count for columns are')
print (' Day     : '+ str(df['day'].isnull().sum()))
print (' Month   : '+ str(df['month'].isnull().sum()))
print (' Year    : '+ str(df['year'].isnull().sum()))
print (' station : '+ str(df['station'].isnull().sum()))
print (' maxtp   : '+ str(df['maxtp'].isnull().sum()))
print (' mintp   : '+ str(df['mintp'].isnull().sum()))
print (' rain    : '+ str(df['rain'].isnull().sum()))
print (' wdsp    : '+ str(df['wdsp'].isnull().sum()))
print (' hg      : '+ str(df['hg'].isnull().sum()))
print (' sun     : '+ str(df['sun'].isnull().sum()))
```

```python
#4. Create a numerical summary (mean, standard deviation, minimum, maximum, etc.) for each
#   of the weather measurements. Discuss and interpret your results.

# Explanation:
#      Describe function displays the statistical data of paramater provided.
#      Below we calculate statistical data of all weather points individually for the whole year.
#      This provides the count of datapoints, Min and max records.
#      Mean provides the average of all datapoints, whereas median (also 50% mentioned below) sorts 
#.     the datapoints and it is middle of the list.
#      These two above variables decide how the data is distributed.
#      (25,50,75%) are the quantiles of datapoints in which the maximum count of data is distributed.
    
#      1. Maxtp,Mintp - density of the datapoints are near to the avergae of max tp values as median and mean 
#        are close.
        
#      2. Rain - mean and median are far away hence this decribes there are more datapoints at the beginning 
#                 and some outliers which fetches the median far from mean.
        
#      3. Wind speed,highest gust,sun - This feature has normal curve but due to some outliers it stretches 
#                                     mean away from median.
     


#Code:
print ("*Stats data for Max Temp :*")
print(df['maxtp'].describe())
print('Median',df['maxtp'].median())

print ("\n*Stats data for Min Temp :*")
print(df['mintp'].describe())
print('Median',df['mintp'].median())

print ("\n*Stats data for Rain :*")
print(df['rain'].describe())
print('Median',df['rain'].median())

print ("\n*Stats data for Wind Speed :*")
print(df['wdsp'].describe())
print('Median',df['wdsp'].median())

print ("\n*Stats data for highest Gust :*")
print(df['hg'].describe())
print('Median',df['hg'].median())

print ("\n*Stats data for sun duration :*")
print(df['sun'].describe())
print('Median',df['sun'].median())
```
